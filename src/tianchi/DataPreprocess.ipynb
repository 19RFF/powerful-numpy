## 数据的维度

### 一维

在 Python 中，一般有以下几种常用的数据结构：

- list：列表。一组元素，可以重复。
- tuple：元组。一组元素，可以重复，但不可变。
- set：集合。一组元素，无重复值。
- dict：字典。一组Key/Value对，Key 不重复。

Python 的元素是不限定类型的，可以是任意类型，甚至函数/类。这些数据结构都可以轻易地表达一维数据。

```python
class A: ...
def a() -> None: ...
```

#### 列表 list

```python
# list
# 元素分别是：整数、浮点数、字符串，类实例、类本身、函数返回值、函数本身，以及后面的——可重复之前的元素
[1, 2.0, "32", A(), A, a(), a, 1, 2.0, "32"]
```

#### 元组 tuple

```python
# 和上面的 list 元素一样
# 你会发现除了包围的括号不一样外，其他好像没任何变化
# 确实，它们最大的不同就是 tuple 具有不可变性
(1, 2.0, "32", A(), A, a(), a, 1, 2.0, "32")
```

下面展示 tuple 的不可变性：

```python
# 展示 tuple 不可变性
tup1 = (2, 3)
lst1 = [2, 3]
```

```python
# 元素不可改变
tup1[0] = 1
```

```python
# 元素可以改变
lst1[0] = 1
lst1
```

```python
# tuple 可哈希
hash(tup1)
```

```python
# list 不可哈希
hash(lst1)
```

>可哈希：如果对象具有在其生存期内从不更改的哈希值（它需要一个 __hash__（） 方法），并且可以与其他对象进行比较（它需要一个 __eq__（） 方法），则该对象是可哈希的。比较相等的可哈希对象必须具有相同的哈希值。  
摘自：https://docs.python.org/3.8/glossary.html

tuple 占用空间也会略小一些：

```python
import sys
```

```python
sys.getsizeof([])
```

```python
sys.getsizeof(())
```

此外，list 更容易删除和插入，tuple 只能创建一个新的；同时 list 有很多内置方法，tuple 却很少。

#### 集合set

```python
# 我们还用之前的数据
# 这次，我们没有重复值了！
set((1, 2.0, "32", A(), A, a(), a, 1, 2.0, "32"))
```

```python
# 即便是类/函数也一样，但——注意，每个实例是不一样的！
# 注意，这里我们换个 set 的使用方式——「用一个大括号」
{A, A, A(), A(), a, a, a(), a()}
```

```python
# 另外，set 的元素一定要是可哈希的
{[]}
```

看看集合这特性，就知道它最适合用来去重了。

#### 小结

列表、元组、集合和字典都可以很方便地表示一维数据：

- 列表最常用，可以方便地进行元素的增、删、改、遍历等操作。
- 元组一般用于表示一组有意义的元素，如坐标（39.9042° N, 116.4074° E），开始和结束的位置（0，3）或用来哈希。
- 集合一般用于元素去重，注意集合是无序的。
- 字典一般用于表示Key/Value对。

另外，我们也总结一下常用的写法（每种方法都可以）：

```python
# 数组：一般用 a 的写法
a = [] 
b = list()
```

```python
# 元组：一般用 a 的写法
a = (2, 3)
b = tuple((2, 3))
```

```python
# 集合：一般用 a 的写法
a = {2, 3}
b = set((2, 3))

# 但空集合一般用 b 的写法
empty_set = set()
```

```python
# 字典：一般用 a 的写法
a = {"name": "Yam", "age": 22}
b = dict({"name": "Yam", "age": 22})

# 空字典一般写作（不是空集合哦）：
empty_dict = {}
```

### 二维

上面提到的数据结构通过嵌套就可以表达多维数据了。但不要忘记，集合不能嵌套，但可以放元组。

#### Python基础数据结构

```python
# 先用列表表示一个
[
    [1, 2], [3, 4], [5, 6]
]
```

```python
# 来个 tuple
(
    (1, 2), (3, 4), (5, 6)
)
```

```python
# set 不可以！
{
    {1, 2}, {3, 4}, {5, 6}
}
```

```python
# 字典
{
    1: {"x": 1, "y": 2},
    2: {"x": 3, "y": 4},
    3: {"x": 5, "y": 6}
}
```

当然我们大多数时候会根据自己的需要灵活使用几种组合。

```python
# list + dict
[
    {"x": 1, "y": 2},
    {"x": 3, "y": 4},
    {"x": 5, "y": 6}
]
```

```python
# dict + other
{
    1: [1, 2],
    2: (3, 4),
    3: {5, 6}
}
```

```python
# list + tuple
[
    (1, 2), (3, 4), (5, 6)
]
```

## 数据的类型

从大的方面来看，可以分为结构化数据、非结构化数据和半结构化数据。

### 结构化数据

对结构化数据，一般可以使用数组+字典处理，或者直接使用 Pandas 之类的工具统一处理，正如前面的例子。

```python
!head data/table.txt
```

```python
tb = pd.read_csv("./data/table.txt", sep="\t")
tb
```

```python
tb.info()
```

或者就自己写读取逻辑：

```python
data = []
with open("./data/table.txt") as f:
    for i, line in enumerate(f):
        # 第一行 header 跳过去
        if i == 0:
            continue
        org, est, ceo = line.strip().split("\t")
        item = {"organization": "org", "established": int(est), "ceo": ceo}
        data.append(item)
```

```python
data
```

不过对于结构化数据（或其他你要处理的「类」）来说，我们一般会先设计其 Schema，正如数据库设计 Schema 一样。

```python
from dataclasses import dataclass
```

```python
@dataclass
class TechCompany:
    
    organization: str
    established: int
    ceo: str
```

```python
companies = []
for item in data:
    tc = TechCompany(**item)
    companies.append(tc)
```

```python
companies
```

### 半结构化数据

半结构化数据一般也有不同的处理工具，比如解析html的BeautifulSoup，解析XML的xml.etree.ElementTree。

#### xml

```python
import xml.etree.ElementTree as ET
tree = ET.parse('./data/country.xml')
root = tree.getroot()
```

```python
for child in root:
    print(child.tag, child.attrib)
    for ele in child:
        print("\t", ele.tag, ele.text)
```

#### html

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(open("./data/index.html"))
```

```python
for href in soup.find_all("a"):
    link = href.get("href")
    title = href.text
    if link and link.startswith("https"):
        print(title, link)
```
